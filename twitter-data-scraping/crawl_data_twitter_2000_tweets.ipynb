```python
import os
import pandas as pd
from datetime import datetime, timedelta

# Twitter Auth Token
twitter_auth_token = '99be5a7df62707ce8059186d6c1efa0c9688ee5d'  # Change this auth token

# Search keyword and other parameters
search_keyword = 'terör'
limit_per_day = 50
start_date = datetime(2025, 1, 1)
end_date = datetime(2025, 1, 3)
output_directory = r'C:\Users\doganeren.ozer\Desktop\twitter scraping\teror'

# Create the output directory if it doesn't exist
os.makedirs(output_directory, exist_ok=True)

# Function to generate date range
def daterange(start_date, end_date):
    for n in range((end_date - start_date).days + 1):
        yield start_date + timedelta(n)

# Iterate through each date in the range
for single_date in daterange(start_date, end_date):
    # Format dates for Twitter search query
    since_date = single_date.strftime("%Y-%m-%d")
    until_date = (single_date + timedelta(1)).strftime("%Y-%m-%d")

    # Define file name for the current day
    filename = f"tweets_{since_date}.csv"
    filepath = os.path.join(output_directory, filename)

    # Twitter scraping command
    search_query = f"{search_keyword} since:{since_date} until:{until_date}"
    command = (
        f"npx -y tweet-harvest@2.6.1 -o \"{filepath}\" -s \"{search_query}\" --tab \"LATEST\" -l {limit_per_day} --token {twitter_auth_token}"
    )

    # Execute the command
    print(f"Running command for date: {since_date}")
    os.system(command)

    # Verify if the file was created and print its contents
    if os.path.exists(filepath):
        print(f"Data saved to {filepath}")
        df = pd.read_csv(filepath, delimiter=",")
        display(df.head())
    else:
        print(f"Failed to fetch data for {since_date}")
```
